```python
sns.countplot(x='Survived', data=df, palette='Set2')
plt.title('Survival Count (0 = Died, 1 = Survived)')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python

sns.countplot(x='Sex', hue='Survived', data=df, palette='Set2')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Age'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
embarked_counts = df['Embarked'].value_counts()
labels = embarked_counts.index.map({'S': 'Southampton', 'C': 'Cherbourg', 'Q': 'Queenstown'})

plt.figure(figsize=(6, 6))
plt.pie(embarked_counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#66b3ff', '#99ff99', '#ffcc99'])
plt.title('Passenger Distribution by Embarkation Port')
plt.axis('equal')
plt.show()
```
