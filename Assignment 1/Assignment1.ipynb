```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Iris.csv")
plt.hist(df['SepalLengthCm'], bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Count")
plt.show()
```

```python

sns.boxplot(x='Species', y='PetalLengthCm', data=df)
plt.title("Petal Length by Species")
plt.show()
```

```python
sns.pairplot(df, hue='Species')
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()
```

```python

sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=df)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Brand': ['Maruti', 'Hyundai', 'Honda', 'Toyota', 'Ford'],
    'Count': [100, 60, 40, 30, 20]
}

df = pd.DataFrame(data)
plt.figure(figsize=(6, 6))
plt.pie(df['Count'], labels=df['Brand'], autopct='%1.1f%%', startangle=140)
plt.title('Car Brand Distribution - Pie Chart')
plt.axis('equal')  
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Brand': ['Maruti', 'Hyundai', 'Honda', 'Toyota', 'Ford'],
    'Count': [100, 60, 40, 30, 20]
}

df = pd.DataFrame(data)
plt.figure(figsize=(8, 5))
plt.bar(df['Brand'], df['Count'], color='skyblue')
plt.title('Car Brand Distribution - Bar Graph')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
